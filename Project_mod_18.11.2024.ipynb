```python
#--1.Find out the top 5 months in terms of impressions?
```

```python
#--Step1.Converting the date format.
df_Table1['Date'] = pd.to_datetime(df_Table1['Date'])
df_Table1
```

```python
#--Step2.Segregating day, month and year columns respectively.
df_Table1['Day'] = df_Table1['Date'].dt.day
df_Table1['month'] = df_Table1['Date'].dt.month
df_Table1['year'] = df_Table1['Date'].dt.year
df_Table1
```

```python
#--Step3.Filtering 'Impressions' value from the 'Metric' column.
df1_Table1 = df_Table1[df_Table1['Metric'] == 'Impressions']
df1_Table1
```

```python
#--Step4.Using groupby extracting data's of 'Impressions' value from the 'Metric' column for all the months and year.
df2_Table1 = (df1_Table1.groupby(['month','Metric','year'])['Value'].sum()).sort_values(ascending = False).reset_index()
df2_Table1
```

```python
#--Step5.Extracting top 5 months in terms of'Impressions'.
df2_Table1.head(5)
```

```python
#--2.Which model has maximum impressions?
```

```python
#--Step1.Merging two tables using primary key.
dfmerge_table = pd.merge(df_Table1, df_Table2 , on ='Campaign')
dfmerge_table
```

```python
#--Step2.Filtering 'Impressions' value from the 'Metric' column.
df3_Table1 = dfmerge_table[dfmerge_table['Metric'] == 'Impressions']
df3_Table1
```

```python
#--Step4.Extracting 'model' for maximum 'Impressions'.
(df3_Table1.groupby(['Model','Metric'])['Value'].sum()).sort_values(ascending = False).reset_index().head(1)
```

```python
#--5.Find out the top 3 campaigns in terms of CPM (CPM stands for cost per mile which means cost per 1000 impressions. 
#--CPM = total media cost/total impressions * 1000)
```

```python
#--Step1.Filtering 'Impressions' value from the 'Metric' column.
df8_Table1 = df_Table1[df_Table1['Metric'] == 'Impressions']
df8_Table1
```

```python
#--Step2.Creating a new column as 'Total Impressions' in terms of 'Impressions' and 'Campaign'.
df9_Table1 = df8_Table1.groupby(['Campaign','Metric'])['Value'].sum().reset_index(name='Total Impressions')
df9_Table1
```

```python
#--Step3.Sorting the 'Total Impressions' values.
df10_Table1 = df9_Table1.sort_values(by='Total Impressions', ascending=False)
df10_Table1
```

```python
#--Step4.Filtering 'Media Cost' value from the 'Metric' column.
df11_Table1 = df_Table1[df_Table1['Metric'] == 'Media Cost']
df11_Table1
```

```python
#--Step5.Creating a new column as 'Total Media Cost' in terms of 'Media Cost' and 'Campaign'.
df12_Table1 = df11_Table1.groupby(['Campaign','Metric'])['Value'].sum().reset_index(name='Total Media Cost')
df12_Table1
```

```python
#--Step6.Sorting the 'Total Media Cost' values.
df13_Table1 = df12_Table1.sort_values(by='Total Media Cost', ascending=False)
df13_Table1
```

```python
#--Step7.Merging 'Total Impressions' and 'Total Media Cost' in terms of 'Campaign'.
dfmerge2_table = pd.merge(df10_Table1 , df13_Table1 , on = 'Campaign')
dfmerge2_table
```

```python
#--Step8.Creating a new column as 'CPM' in terms 'Campaign'.
dfmerge2_table['CPM'] = (dfmerge2_table['Total Media Cost'] / dfmerge2_table['Total Impressions']) * 1000
```

```python
#--Step9.Filtering top 3 'Campaign' in terms of 'CPM'. 
dfmerge2_table.sort_values(by = 'CPM', ascending = False).head(3)
```

```python
#--7.Which media type is more expensive in terms of CPM?
```

```python
#--Step1.Merging two tables using primary key.
dfmerge3_table = pd.merge(df_Table1, df_Table2 , on ='Campaign')
dfmerge3_table
```

```python
#--Step2.Checking for the unique values in 'Media Type'.
dfmerge3_table['Media Type'].unique()
```

```python
#--Step3.Filtering 'Media Cost' value from the 'Metric' column.
df15_Table1 = dfmerge3_table[dfmerge3_table['Metric'] == 'Media Cost']
df15_Table1
```

```python
#--Step4.Creating a new column as 'Total Media Cost' in terms of 'Media Type'.
df16_Table1 = df15_Table1.groupby(['Media Type','Metric'])['Value'].sum().reset_index(name='Total Media Cost')
df16_Table1
```

```python
#--Step5.Filtering 'Impressions' value from the 'Metric' column.
df17_Table1 = dfmerge3_table[dfmerge3_table['Metric'] == 'Impressions']
df17_Table1
```

```python
#--Step6.Creating a new column as 'Total Impressions' in terms of 'Media Type'.
df18_Table1 = df17_Table1.groupby(['Media Type','Metric'])['Value'].sum().reset_index(name='Total Impressions')
df18_Table1
```

```python
#--Step7.Merging 'Total Impressions' and 'Total Media Cost' in terms of 'Media Type'.
dfmerge4_table = pd.merge(df16_Table1 , df18_Table1 , on = 'Media Type')
dfmerge4_table
```

```python
#--Step8.Creating a new column as 'CPM' in terms 'Media Type'.
dfmerge4_table['CPM'] = (dfmerge4_table['Total Media Cost'] / dfmerge4_table['Total Impressions']) * 1000
dfmerge4_table
```

```python
#--Step9.Sorting the 'CPM' values to find out the expensive 'Media Type'.
dfmerge4_table.sort_values(by='CPM', ascending=False)[:1]
```
